```python
def _post_process_cor(autocorr):
    autocorr = np.array(autocorr)
    autocorr[np.isinf(autocorr)] = np.nan
    means = np.nanmean(autocorr,axis=0)
    stderr = np.nanstd(autocorr,axis=0) / np.sqrt(np.sum(~np.isnan(autocorr), axis=0))
    return means, stderr
```

## C(r) analysis

```python
## Define atac particles with high/low H3K27me3 signal
atac_peaks = pd.read_table("../1_preprocess/atacpeak_score50.bed",header=None)
atac_peaks.columns = ["chrom","start","end"]
atac_peaks["pos"] =  ((atac_peaks["start"] + atac_peaks["end"]) / 2 // 5000 * 5000).astype(int)
atac_peaks = atac_peaks.groupby(["chrom","pos"]).size().reset_index(name="count")
chroms = ["chr"+str(i) for i in range(1,20)] + ["chrX"]
atac_peaks = atac_peaks.query("chrom in @chroms")

# load fragments 
ct_frags = pd.read_table("../1_preprocess/pseudobulk/ct.fragments.gz",header=None).iloc[:,:3]
atac_frags = pd.read_table("../1_preprocess/pseudobulk/atac.fragments.gz",header=None).iloc[:,:3]
ct_frags.columns = ["chrom","start","end"]
atac_frags.columns = ["chrom","start","end"]
ct_frags["pos"] =  ((ct_frags["start"] + ct_frags["end"]) / 2 // 5000 * 5000).astype(int)
atac_frags["pos"] =  ((atac_frags["start"] + atac_frags["end"]) / 2 // 5000 * 5000).astype(int)
ct_frags = ct_frags.groupby(["chrom","pos"]).size().reset_index(name="count")
atac_frags = atac_frags.groupby(["chrom","pos"]).size().reset_index(name="count")
```

```python
merged = atac_peaks.merge(atac_frags, on=["chrom","pos"], how="left",suffixes=("", "_atac")).fillna(0)
merged = merged.merge(ct_frags, on=["chrom","pos"], how="left",suffixes=("", "_ct")).fillna(0)
```

```python
atac_peaks_particle_type = merged.copy()[["chrom","pos","count_atac","count_ct"]]
scaler = StandardScaler()
atac_peaks_particle_type["norm_count_atac"] = scaler.fit_transform(atac_peaks_particle_type["count_atac"].values.reshape(-1, 1))
atac_peaks_particle_type["norm_count_ct"] = scaler.fit_transform(atac_peaks_particle_type["count_ct"].values.reshape(-1, 1))

atac_peaks_particle_type["particle_type"] = np.where(atac_peaks_particle_type["norm_count_atac"] > atac_peaks_particle_type["norm_count_ct"], "Acc_High", "H3K27me3_High")
```

```python
def _load_cell(cellname, resolution):
    cell = Cell3D.Cell3D(
        cellname = cellname,
        resolution = int(resolution),
        tdg_path = f"../data/tdg/{cellname}.5k.3dg.gz",
    )
    cell.add_bed_data(path=f"../data/fragments/atac_frags/{cellname}.atac.frag.bed.gz", 
                     column_name="atac" ,keep_3prime=True,type="allelic_resolved")
    cell.add_bed_data(path=f"../data/fragments/ct_frags/{cellname}.ct.frag.bed.gz", 
                     column_name="ct" ,keep_3prime=True,type="allelic_resolved")
    
    cell.tdg = cell.tdg.merge(atac_peaks_particle_type[["chrom","pos","particle_type"]],how="outer")
    cell.tdg["particle_type"] = cell.tdg["particle_type"].fillna("Else")
    cell.add_RNA_data(rnag1 = rnag1, rnag2 = rnag2, genes = genes,type = "tss")

    cell.subset(query='atac > 0 or ct > 0 or UMI >= 1',in_place=True)
    cell.tdg = cell.tdg.dropna()
    
    return cell

def _load_cell_allpoints(cellname,resolution):
    cell = Cell3D.Cell3D(
        cellname = cellname,
        resolution = int(resolution),
        tdg_path = f"../data/tdg/{cellname}.5k.3dg.gz",
    )

    cell.add_chrom_length(chrom_length_path = "/share/Data/public/ref_genome/mouse_ref/M23/raw_data/chr.dip.len")

    cell.calc_radial_position(if_rank=True,if_norm_mean=False)

    # cell.add_bedGraph_data(
    #     path=f"/share/Data/public/ref_genome/mouse_ref/M23/CpG/normal_cpg/M23.CpG.5000.txt",
    #     column_name="CpG",
    #     type="all"
    # )
    #cell.add_feature_in_radius(feature="CpG",radius=2,type="sum",if_self=True,if_rank=True)
    return cell

with concurrent.futures.ProcessPoolExecutor(max_workers=128) as executor:
    cells = list(tqdm.tqdm(executor.map(_load_cell, metadata['cellname'], [5000]*len(metadata)), total=len(metadata)))
    cells_allpoints = list(tqdm.tqdm(executor.map(_load_cell_allpoints, metadata['cellname'], [5000]*len(metadata)), total=len(metadata)))
```

```python
def _calc_spatial_crosscorr(cell,query1,query2):
    np.random.seed(42)
    df = cell.get_data()
    points1 = df.query(query1)[["x","y","z"]].values.copy()
    points2 = df.query(query2)[["x","y","z"]].values.copy()
    if points1.shape[0] > 2000:
        points1 = points1[np.random.choice(points1.shape[0], 2000, replace=False), :]
    else:
        return np.array([np.nan for i in range(10)])
    if points2.shape[0] > 2000:
        points2 = points2[np.random.choice(points2.shape[0], 2000, replace=False), :]
    else:
        return np.array([np.nan for i in range(10)])

    return spatialstat.spatial_crosscorr_3d(points1,points2,dr=1,max_dist=10,Nsim=4)[0]


with concurrent.futures.ProcessPoolExecutor(max_workers=128,mp_context=multiprocessing.get_context("fork")) as executor:
    atac_ct = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0',query2='ct > 0'),cells), total=len(cells)))
    ct_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='ct > 0',query2='UMI >= 1'),cells), total=len(cells)))
    atac_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0',query2='UMI >= 1'),cells), total=len(cells)))
    random_random = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='chrom_code >= 0',query2='chrom_code >= 0'),cells_allpoints), total=len(cells_allpoints)))

    me3low_atac_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0 & particle_type == "Acc_High"',query2='UMI >= 1'),cells), total=len(cells)))
    me3high_atac_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0 & particle_type == "H3K27me3_High"',query2='UMI >= 1'),cells), total=len(cells)))

    me3low_atac_me3 = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0 & particle_type == "Acc_High"',query2='ct > 0'),cells), total=len(cells)))
    me3high_atac_me3 = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0 & particle_type == "H3K27me3_High"',query2='ct > 0'),cells), total=len(cells)))


means_atac_ct, stderr_atac_ct = _post_process_cor(atac_ct)
means_ct_rna, stderr_ct_rna = _post_process_cor(ct_rna)
means_atac_rna, stderr_atac_rna = _post_process_cor(atac_rna)
means_random_random, stderr_random_random = _post_process_cor(random_random)

means_me3low_atac_rna, stderr_me3low_atac_rna = _post_process_cor(me3low_atac_rna)
means_me3high_atac_rna, stderr_me3high_atac_rna = _post_process_cor(me3high_atac_rna)

means_me3low_atac_me3, stderr_me3low_atac_me3 = _post_process_cor(me3low_atac_me3)
means_me3high_atac_me3, stderr_me3high_atac_me3 = _post_process_cor(me3high_atac_me3)
```

```python
def _post_process_cor(autocorr):
    autocorr = np.array(autocorr)
    autocorr[np.isinf(autocorr)] = np.nan
    means = np.nanmean(autocorr,axis=0)
    stderr = np.nanstd(autocorr,axis=0) / np.sqrt(np.sum(~np.isnan(autocorr), axis=0))
    return means, stderr

means_atac, stderr_atac = _post_process_cor(cor_atac)
#means_ct, stderr_ct = _post_process_cor(cor_ct)
# means_atac_next, stderr_atac_next = _post_process_cor(cor_atac_next)
# means_ct_next, stderr_ct_next = _post_process_cor(cor_ct_next)
means_atac_random, stderr_atac_random = _post_process_cor(cor_atac_random)
#means_ct_random, stderr_ct_random = _post_process_cor(cor_ct_random)

means_all, stderr_all = _post_process_cor(cor_all)
```

```python
def _calc_spatial_crosscorr(cell,query1,query2):
    np.random.seed(42)
    df = cell.get_data()
    points1 = df.query(query1)[["x","y","z"]].values.copy()
    points2 = df.query(query2)[["x","y","z"]].values.copy()
    if points1.shape[0] > 2000:
        points1 = points1[np.random.choice(points1.shape[0], 2000, replace=False), :]
    else:
        return np.array([np.nan for i in range(10)])
    if points2.shape[0] > 2000:
        points2 = points2[np.random.choice(points2.shape[0], 2000, replace=False), :]
    else:
        return np.array([np.nan for i in range(10)])

    return spatialstat.spatial_crosscorr_3d(points1,points2,dr=1,max_dist=10,Nsim=4)[0]


with concurrent.futures.ProcessPoolExecutor(max_workers=128,mp_context=multiprocessing.get_context("fork")) as executor:
    atac_ct = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0',query2='ct > 0'),cells), total=len(cells)))
    ct_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='ct > 0',query2='rna >= 1'),cells), total=len(cells)))
    atac_rna = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='atac > 0',query2='rna >= 1'),cells), total=len(cells)))
    random_random = list(tqdm.tqdm(executor.map(partial(_calc_spatial_crosscorr,query1='chrom_code >= 0',query2='chrom_code >= 0'),cells_allpoints), total=len(cells_allpoints)))

means_atac_ct, stderr_atac_ct = _post_process_cor(atac_ct)
means_ct_rna, stderr_ct_rna = _post_process_cor(ct_rna)
means_atac_rna, stderr_atac_rna = _post_process_cor(atac_rna)
means_random_random, stderr_random_random = _post_process_cor(random_random)
```
